### Abstract

Improve Autism Screening by creating predicting the likelihood of having this condition.

### About this dataset

What is Autism\
Autism, or autism spectrum disorder (ASD), refers to a broad range of conditions characterized by challenges with social skills, repetitive behaviors, speech and nonverbal communication.

### Causes and Challenges

It is mostly influenced by a combination of genetic and environmental factors. Because autism is a spectrum disorder, each person with autism has a distinct set of strengths and challenges. The ways in which people with autism learn, think and problem-solve can range from highly skilled to severely challenged.\
Research has made clear that high quality early intervention can improve learning, communication and social skills, as well as underlying brain development. Yet the diagnostic process can take several years.

### The Role of Machine Learning

This dataset is composed of survey results for more than 700 people who filled an app form. There are labels portraying whether the person received a diagnosis of autism, allowing machine learning models to predict the likelihood of having autism, therefore allowing healthcare professionals prioritize their resources.

Columns
-------

-   `ID` - ID of the patient
-   `A1_Score to A10_Score` - Score based on Autism Spectrum Quotient (AQ) 10 item screening tool
-   `age` - Age of the patient in years
-   `gender` - Gender of the patient
-   `ethnicity` - Ethnicity of the patient
-   `jaundice` - Whether the patient had jaundice at the time of birth
-   `autism` - Whether an immediate family member has been diagnosed with autism
-   `contry_of_res` - Country of residence of the patient
-   `used_app_before` - Whether the patient has undergone a screening test before
-   `result` - Score for AQ1-10 screening test
-   `age_desc` - Age of the patient
-   `relation` - Relation of patient who completed the test
-   `Class/ASD` - Classified result as 0 or 1. Here 0 represents No and 1 represents Yes. This is the target column, and during submission submit the values as 0 or 1 only.

**Data Cleaning**

The data which is obtained from the primary sources is termed the raw data and required a lot of preprocessing before we can derive any conclusions from it or to some modelling on it. Those preprocessing steps are known as **data cleaning** and it includes, outlier removal, null value imputation, and removing discrepancies of any sort in the data inputs..ts.

```python
def fix_spell_mistake(df):
    df.rename(columns={"austim":"autism"},inplace=True) #Spelling Mistake
    df.rename(columns={"contry_of_res":"country_of_res"},inplace=True) #Spelling Mistake
fix_spell_mistake(df)
```

```python
def feature_count(data):
    # missing values
    total_missing = data.isnull().sum().sort_values(ascending = False)
    percent_missing = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    missing = pd.concat([total_missing, percent_missing], axis=1, keys=['Total', 'Percent'])
    
    #unique values 
    total_data = data.count()
    tt = pd.DataFrame(total_data)
    tt.columns = ['Total']
    uniques = []
    for col in data.columns:
        unique = data[col].nunique()
        uniques.append(unique)
    tt['Uniques'] = uniques
    unique = tt 
    return pd.concat([missing, unique], axis=1, keys=['Missing', 'UNIQUE'])
feature_count(df)
```

Feature Engineering
-------------------

Feature Engineering helps to derive some valuable features from the existing ones. These extra features sometimes help in increasing the performance of the model significantly and certainly help to gain deeper insights into the data.

```python
# This functions make groups by taking
# the age as a parameter
def convertAge(age):
	if age < 4:
		return 1
	elif age < 12:
		return 2
	elif age < 18:
		return 3
	elif age < 40:
		return 4
	else:
		return 5

df['ageGroup'] = df['age'].apply(convertAge)
sb.countplot(x=df['ageGroup'].astype(str), hue=df['Class/ASD'].astype(str))
plt.show()
```

```python
def add_feature(data):

    # Creating a column with all values zero
    data['sum_score'] = 0
    for col in data.loc[:,'A1_Score':'A10_Score'].columns:
    	
    	# Updating the 'sum_score' value with scores
    	# from A1 to A10
    	data['sum_score'] += data[col]
    
    # Creating a random data using the below three columns
    data['ind'] = data['autism'] + data['used_app_before'] + data['jaundice']
    
    #return data

add_feature(df)
sb.countplot(x=df['sum_score'], hue=df['Class/ASD'].astype(str))
plt.show()
```

- We had noticed that 'result' and 'age' both the features had skewed distributions which we shall fix by applying log transformations

```python
# Applying log transformations to remove the skewness of the data.
def remove_skew(df):
    df['age']=df['age'].apply(lambda x: np.log(x))
    df['result']=df['result'].apply(lambda x: np.log(x+50))
remove_skew(df)
sb.distplot(df['age'])
plt.show()
```

- Who fills the survey shouldn't affect the diagnosis so we will drop the feature 'relation'
- We will apply one hot encoding on all the categorical features as labels can confuse the model

```python
X=df.drop(['ID','Class/ASD','relation'],axis=1)
X=pd.get_dummies(X)
Y=df['Class/ASD']
X.shape
```

- We already have a test set but we shall create a cross validation set to judge the models and pick one based on cross calidation error
- We shall artificially increase the positive examples to improve the models ability to diagnose autism. This is important as the number of positive examples are very less as compared to negative

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size = 0.2, random_state=10)

# As the data was highly imbalanced we will balance it by adding repetitive rows of minority class.
ros = RandomOverSampler(sampling_strategy='minority',random_state=42)
X_train, Y_train = ros.fit_resample(X_train,Y_train)
X_train.shape, Y_train.shape
```

```python
# Normalizing the features for stable and fast training.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
new_models = [LogisticRegression(), XGBClassifier(), SVC(kernel='rbf')]

for model in new_models:
    model.fit(X_train, Y_train)

    print(f'{model} : ')
    print('Training Accuracy : ', metrics.roc_auc_score(Y_train, model.predict(X_train)))
    print('Validation Accuracy : ', metrics.roc_auc_score(Y_val, model.predict(X_val)))
    print()
```

```python
test_df=pd.read_csv('./data/test.csv')
X_test=test_df.drop(['ID'],axis=1)
def transform(df):
    fix_spell_mistake(df)
    clean(df)
    df['ageGroup'] = df['age'].apply(convertAge)
    add_feature(df)
    remove_skew(df)
    df.drop(['relation'],axis=1,inplace=True)
    df=pd.get_dummies(df)
    for c in X.columns:
        if c not in df.columns:
            df[c]=False
    for c in df.columns:
        if c not in df.columns:
            df.drop([c],axis=1,inplace=True)
    df=df[X.columns]
    df=scaler.transform(df)
    return df
X_test=transform(X_test)
Y=new_models[1].predict(X_test)
```

```python
ans=pd.DataFrame(Y,columns=['Class/ASD'])
ans.set_index(test_df['ID'],inplace=True)
ans.to_csv("./ans.csv")
```
